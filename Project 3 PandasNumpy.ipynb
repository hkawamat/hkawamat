We would need to make the 'price' and 'odometer' series into 'int', to do this we must replace all of the strings before we convert the string into an int. 

```python
autos.loc[:, 'price'] = autos.loc[:, 'price'].str.replace(',','').str.replace('$', '').astype(int)
```

```python
autos.loc[:, 'odometer'] = autos.loc[:, 'odometer'].str.replace('km', '').str.replace(',',  '').astype(int)
```

After we have done this, we use the info() function to check whether or not 'price' and 'odometer' has changed from an object to a string. 

```python
autos.info()
```

We should rename the 'odometer' column label to 'odometer_km' since we replaced the 'km' to make this column an int.

```python
autos.rename({'odometer': 'odometer_km'}, axis = 1 ,inplace = True)
```

Choose brands that represent more than 2% of the sales, which would mean that Fiat would be last car brand included in our list.

```python
auto_brand_specific = autos['brand'].value_counts(normalize = True).head(9)
```

```python
auto_brand_specific.index
```

```python
index_brand = auto_brand_specific.index
```

We are going to create a dictionary with the mean prices of each brand

```python
brand_mean_list = {}
```

```python
for a in index_brand:
    brand_data = autos.loc[autos['brand'] == a, :]
    mean_of_brand = brand_data['price'].mean()
    brand_mean_list[a] = mean_of_brand
```

```python
brand_mean_list
```

For each brand within the Top 9, we have created a dictionary with the mean prices of the cars. We can see that Audi has the most expensive cars and Renault has the least expensive cars in terms of the mean prices. 

```python
new_brand_mean_list = sorted(brand_mean_list.items(), key = lambda kv: kv[1], reverse = True)
```

```python
new_brand_mean_list
```

```python
mean_mileage = {}
for a in index_brand:
    brand_list = autos.loc[autos['brand'] == a, :]
    mileage_mean = brand_list['odometer_km'].mean()
    mean_mileage[a] = mileage_mean
```

```python
mean_mileage
```

```python
mean_price_series = pd.Series(brand_mean_list)
```

```python
mean_price_dataframe = pd.DataFrame(mean_price_series, columns = ['Mean_Price'])
```

```python
mean_mileage_series = pd.Series(mean_mileage)
```
