### Rename Columns and rename values by mapping

Use .rename and .map to rename values and columns

```python
# Mapping specific answers to boolean
new_map = { "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True}
    
# To rename the titles, create a dictionary that helps rename the current titles to new titles
new_name = {'Which of the following Star Wars films have you seen? Please select all that apply.': 'seen_1',
            'Unnamed: 4': 'seen_2',
            'Unnamed: 5': 'seen_3',
            'Unnamed: 6': 'seen_4',
            'Unnamed: 7': 'seen_5',
            'Unnamed: 8': 'seen_6'}
```

```python
# Rename the columns to the new columns
star_wars = star_wars.rename(columns = new_name)
```

```python
# Map each column by changing each answer to boolean/ NaN
# Change each column to float
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(new_map)
    star_wars[col] = star_wars[col].astype(float)
```

```python
# Display new columns and values
star_wars
```

```python
# Rename anotehr part of the columns to new columns
newer_names = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'ranking_1',
               'Unnamed: 10': 'ranking_2',
               'Unnamed: 11': 'ranking_3',
               'Unnamed: 12': 'ranking_4',
               'Unnamed: 13': 'ranking_5',
               'Unnamed: 14': 'ranking_6'}
```

```python
# Rename the columns to new columns
star_wars = star_wars.rename(columns = newer_names)
```

```python
# Display the table with the new named columns
star_wars
```

### Calculate and Plot the mean of the rankings

We calculate the mean of the specific columns and plot them on a graph

```python
# Get the mean of the rankings columns to decide which movie had the best ranking
rankings = star_wars[star_wars.columns[9:15]].mean()
```

```python
# Plot the mean of the rankings columns
rankings.plot.bar()
```
