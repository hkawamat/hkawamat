### Create New Column 'dissatisfied'  
  The dissatisifed column consists of True, False, NaN values that indicates whether the employee resigned due to dissatisfaction

```python
# List Frequency Table of the Dissatifaction Column
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
# List Frequency Table of the Job Dissatifaction Column
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
# Function which intakes the each element and returns either False, NaN, or True.
def update_vals(series):
        if series == '-':
            return False
        elif pd.isnull(series):
            return np.nan
        else:
            return True
     
```

```python
# Assign the dissatisfied column for Tafe which holds true if any one element of the 2 is true.
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(update_vals).any(axis = 1, skipna = False)
```

```python
# Frequency List of Dissatisfied while also accounting for Nulls
tafe_resignations_up = tafe_resignations.copy()
tafe_resignations_up['dissatisfied'].value_counts(dropna = False)
```

```python
# Assign the dissatisfied column for Tafe which holds true if any one element of the multiple elements is true.
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction', 'dissatisfaction_with_the_department', 'physical_work_environment', 'lack_of_recognition', 'lack_of_job_security', 'work_location', 'employment_conditions', 'work_life_balance', 'workload']].any(1, skipna=False)
dete_resignations_up = dete_resignations.copy()
```

```python
# Frequency List of Dissatisfied while also accounting for Nulls
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

### Create a Column 'Institute', Merge Both Tables, and Drop Columns
  
  Create a column that labels each table as 'DETE' or 'TAFE'. Also merge the two tables and label as combined and drop certain columns

```python
# Creating a Column called Institute which inidcates whether the row is DETE or TAFE
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
# Concatenate both tables by stacking it on top of each other while maintaining similar columns
combined = pd.concat([dete_resignations_up, tafe_resignations_up], axis = 0, ignore_index = True)
combined.notnull().sum()
```

```python
# Assign the combined table and drop columns with 500 or less non null values
combined_updated = combined.dropna(axis = 1, thresh = 500).copy()
combined_updated.notnull().sum()
```

We have succesfully merged our two tables into one and deleted any columns within the combined table where the columns have 500 or less non null values

### Extract Values and Categorize Values

  We will extract the values by making every value a string then a float, afterwards we will change the float into umbrella terms (strings) that will consist of certain floats for each term

```python
# Frequency List of Each Value

combined_updated['institute_service'].value_counts()
```

```python
# Extracting the first digit from each string, then categorizing the first digit as the float values
combined_updated['skr']= combined_updated['institute_service'].astype(str).str.extract(r'(\d+)').astype(float)
```

```python
# Frequency List of Updated Values
combined_updated['skr'].value_counts()
```

```python
# Function to intake each element and returning new values/ labels (Umbrella Terms)
def numbers_d(element):
    if element < 3.0:
        return 'New'
    elif 6.0 >= element >= 3.0:
        return 'Experienced'
    elif 10.0 >= element >= 7.0:
        return 'Established'
    elif pd.isnull(element):
        return np.nan
    else:
        return 'Veteran'

   
```

```python
# Applying the Function to the 'skr' column and assigning results to 'service_cat' column
combined_updated['service_cat'] = combined_updated['skr'].apply(numbers_d)
```

```python
# Frequency List of Updated Values
combined_updated['service_cat'].value_counts(dropna = False).sort_values(ascending = False)
```

We have made a new column with the categorization/ values that indicate the duration of the employee's service

### Plotting the Dissatisfaction of Employees  
  We will start by changing NaN to False, then we will create a pivot table on the rate of dissatisfaction amongst varying employees, then we will visualize this pivot table to easier understand this insight 

```python
# Frequency List of 'Dissatisfied' Column's Values
combined_updated['dissatisfied'].value_counts(dropna = False)
```

```python
# Change NaN to False
combined_updated['dissatisfied'] = combined_updated['dissatisfied'].fillna(False)
combined_updated['dissatisfied'].value_counts(dropna = False)
```

```python
# Create a Pivot Table based on dissatisfaction amongst varying employees
dissatisfied_pivots = pd.pivot_table(combined_updated, index = 'service_cat', values = 'dissatisfied')
dissatisfied_pivots
```

```python
%matplotlib inline
```

```python
# Plot the Pivot Table on Dissatisfaction
dissatisfied_pivots.plot(kind = 'bar', rot = 360)
```

### Conclusion
  We can conclude that Veterans and Established workers tend to resign due to dissatisfaction more often than Experienced and New workers.
